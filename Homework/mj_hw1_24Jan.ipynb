$f(x)=\frac{\pi}{2}+2^{*}(sin(x)+sin(3x)/3+ ... +sin(nx)/n)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xmin = -np.pi
xmax = np.pi
n_point = 100
dx = (xmax-xmin)/n_point
x = np.arange(xmin,xmax+dx,dx)

def square(f,n,x):
    for iii in range(1,n,2):
        f += 2*np.sin(iii*x)/iii
    return f
   
plt.plot(x,square(np.pi/2,11,x),label='n = 11')
plt.plot(x,square(np.pi/2,21,x),label='n = 21')
plt.plot(x,square(np.pi/2,31,x),label='n = 31')
plt.hlines(np.pi/2,xmin,xmax)
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
